```python
# orange image
max_size = 300
#print(orange.thumbnail((max_size, max_size)))
orange_image = orange.convert('RGB')
color_counts = {}
pixels_orange = orange_image.load() #RGB values extracted from the image

for i in range(orange_image.width):
    for j in range(orange_image.height):
        color = pixels_orange[i, j]
        if color in color_counts:
            color_counts[color] += 1
        else:
            color_counts[color] = 1
sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
num_major_colors = 6
major_colors = [color for color, count in sorted_colors[:num_major_colors]]
print(major_colors) # top 6 rgb values that occur within the image
pixelrgb_orange = np.array(orange_image) # array of all rgb values of each pixel
height, width, _ = pixelrgb_orange.shape

red_image = red.convert('RGB')
pixelrgb_red = np.array(red_image) # array of all rgb values of each pixel
pixels_red = red_image.load() #RGB values extracted from the image

turq_image = turq.convert('RGB')
pixelrgb_turq = np.array(turq_image) # array of all rgb values of each pixel
pixels_turq = turq_image.load() #RGB values extracted from the image

blue_image = blue.convert('RGB')
pixelrgb_blue = np.array(blue_image) # array of all rgb values of each pixel
pixels_blue = blue_image.load() #RGB values extracted from the image

egg_image = egg.convert('RGB')
pixelrgb_egg = np.array(egg_image) # array of all rgb values of each pixel
pixels_egg = egg_image.load() #RGB values extracted from the image
```

```python
# 3D Scatterplot of orange
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from PIL import Image

ax = plt.axes(projection = '3d')

x = []
y = []
z = []
c = []

for row in range(0,orange.height):
  for col in range(0, orange.width):
    pix = pixels_orange[col,row]
    newCol = (pix[0] / 255, pix[1] / 255, pix[2] / 255)

    if(not newCol in c):
      x.append(pix[0])
      y.append(pix[1])
      z.append(pix[2])
      c.append(newCol)

ax.scatter(x,y,z, c = c)
ax.set_title('3D Scatter Plot of Orange')
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image

x = []
y = []
z = []
c = []

# Iterate over the image pixels
for row in range(orange.height):
    for col in range(orange.width):
        pix = pixels_orange[col, row]
        newCol = (pix[0] / 255, pix[1] / 255, pix[2] / 255)

        if newCol not in c:
            x.append(pix[0])
            y.append(pix[1])
            z.append(pix[2])
            c.append(newCol)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=c, alpha = 0.5, s = 0.1)
ax.set_title('3D Scatter Plot of Orange')
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

# Create 2D projections
fig2d_xy = plt.figure()
ax2d_xy = fig2d_xy.add_subplot(111)
ax2d_xy.scatter(x, y, c=c, s = 0.1)
ax2d_xy.set_title('Red and Green Projection')
ax2d_xy.set_xlabel('Red')
ax2d_xy.set_ylabel('Green')

fig2d_xz = plt.figure()
ax2d_xz = fig2d_xz.add_subplot(111)
ax2d_xz.scatter(x, z, c=c, s = 0.1)
ax2d_xz.set_title('Red and Blue Projection')
ax2d_xz.set_xlabel('Red')
ax2d_xz.set_ylabel('Blue')

fig2d_yz = plt.figure()
ax2d_yz = fig2d_yz.add_subplot(111)
ax2d_yz.scatter(y, z, c=c, s = 0.1)
ax2d_yz.set_title('Green and Blue Projection')
ax2d_yz.set_xlabel('Green')
ax2d_yz.set_ylabel('Blue')

plt.show()
```
